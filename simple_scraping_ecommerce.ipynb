#### Scraping HTML from E-commerce

```python
from bs4 import BeautifulSoup
from requests import get

response = get("https://www.madville.com.br/", headers={
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:69.0) Gecko/20100101 Firefox/69.0',
    })

formatted_response = " ".join(response.text.split()).replace('> <', '><')

soup = BeautifulSoup(formatted_response, 'html.parser')
```

#### Reading the necessary data and making it available in CSV

```python
products = soup.find_all("div", {"class": "item-wrapper"})

for product in products:
    id_product = product.find('input', {'id' : 'id-product'})['value']
    description = product.find('h2').get_text()
    price = product.find('span', {'class' : 'price'}).get_text().replace("R$","")
    
    print(id_product + ";" + description + ";" + price + ";")
```
